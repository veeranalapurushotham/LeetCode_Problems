***ğŸ”µ7.Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.
Assume the environment does not allow you to store 64-bit integers (signed or unsigned).
Example 1:
Input: x = 123
Output: 321
Example 2:
Input: x = -123
Output: -321
Example 3:
Input: x = 120
Output: 21***

```python
class Solution:
        
    def reverse(self, x: int) -> int:
        if  x>=-2**31 and x<=2**31-1 :
            n=abs(x)
            k=0
            while(n>0):
                k=k*10+n%10
                n=n//10
            if not (k>=-2**31 and k<=2**31-1):
                return 0
            elif x<0:
                return -k
            else:
                return k
        else:
            return 0
obj=Solution()
print(obj.reverse(-123))
```
